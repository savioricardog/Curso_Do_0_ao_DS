## 1. Qual é a média de preço de compra dos imóveis por 'Nível'?
       - Nivel 0 -> Preço entre R$ 0 e R$ 321.950;
       - Nivel 1 -> Preço entre R$ 321.951 e R$ 450.000;
       - Nivel 2 -> Preço entre R$ 450.000 e R$ 645.000;
       - Nivel 3 -> Acima de R$ 645.000.

```python
# Import clean DataSet 
data = pd.read_csv('D:\\repositorio\\datasets\\kc_house_data.csv')

# Create params
data['level'] = data['price'].apply( lambda x: 0 if x < 321950 else
                                               1 if (x > 321951) & (x < 450000) else
                                               2 if (x > 450000) & (x < 645000) else 3)

# Viewing DataSet after define params 
data.sample(10)
```

```python
# Grouping params and getting result
df = data[['price','level']].groupby('level').mean().reset_index()

# Viewing result
df
```

### Exercise

```python
data = pd.read_csv('D:\\repositorio\\datasets\\kc_house_data.csv')

data['place_id'] = 'na'
data['osm_type'] = 'na'
data['country'] = 'na'
data['country_code'] = 'na'

geolocator = Nominatim(user_agent = 'geopyExercises')

# Loop in DataSet
for i in range (len(data.sample(25))):
    
    # Make query
    query = str(data.loc[i, 'lat']) + ',' + str(data.loc[i, 'long'])
    
    # API request
    response = geolocator.reverse(query)

    # FILL LINES
    if 'place_id' in response.raw:
        data.loc[i, 'place_id'] = response.raw['place_id']
        
    if 'osm_type' in response.raw:
        data.loc[i, 'osm_type'] = response.raw['osm_type']
        
    if 'country' in response.raw['address']:
        data.loc[i, 'country'] = response.raw['address']['country']
        
    if 'country_code' in response.raw['address']:
        data.loc[i, 'country_code'] = response.raw['address']['country_code']
        
data.head(10)
```

## ##  4. Adicione os seguintes filtros ao mapa:
       - Tamanho mínimo da área da sala de estar ✔
       - Número mínimo de banheiros✔
       - Valor máximo do preço ✔
       - Tamanho máximo da área do porão ✔

       - Filtro por ano de contrução.

```python
import ipywidgets as wd
from ipywidgets import interact, interactive, fixed, interact_manual
import plotly.express as px
from matplotlib import gridspec as gs
```

```python
# Make Parameter
data['is_waterfront'] = data['waterfront'].apply( lambda x: 'yes' if x == 1 else 'no') 

# Define params from interactive buttons
price_limit = wd.IntSlider(
    value = int(data['price'].mean()),
    min = int(data['price'].min()),
    max = int(data['price'].max()),
    step = 1,
    description = 'Max value',
    disable = False,
    style = {"description_width": 'initial'})

living_limit = wd.IntSlider(
    value = int(data['sqft_living'].mean()),
    min = data['sqft_living'].min(),
    max = data['sqft_living'].max(),
    step = 1,
    description = 'Min Living room size',
    disable = False,
    style = {"description_width": 'initial'})

baths_limit = wd.IntSlider(
    value = data['bathrooms'].mean(),
    min = data['bathrooms'].min(),
    max = data['bathrooms'].max(),
    step = 1,
    description = 'Min baths values',
    disable = False,
    style = {"description_width": 'initial'})

base_limit = wd.IntSlider(
    value = data['sqft_basement'].mean(),
    min = data['sqft_basement'].min(),
    max = data['sqft_basement'].max(),
    step = 1,
    description = 'Max basement values',
    disable = False,
    style = {"description_width": 'initial'})

condition_limit = wd.IntSlider(
    value = data['condition'].mean(),
    min = data['condition'].min(),
    max = data['condition'].max(),
    step = 1,
    description = 'House condition',
    disable = False,
    style = {"description_width": 'initial'})

const_limit = wd.IntSlider(
    value = data['yr_built'].mean(),
    min = data['yr_built'].min(),
    max = data['yr_built'].max(),
    step = 1,
    description = 'House built',
    disable = False,
    style = {"description_width": 'initial'})
```

```python
# Define map functions
def update_map(df,price_limit, living_limit, baths_limit, base_limit, condition_limit, const_limit):
    # Filter data
    houses = df[(df['price'] < price_limit) & 
               (df['sqft_living'] > living_limit ) & 
               (df['bathrooms'] > baths_limit ) &
               (df['sqft_basement'] < base_limit ) &
               (df['condition'] == condition_limit ) &
               (df['yr_built'] > const_limit )] [['id','lat','long','price']].copy()


    # Plot map
    fig = px.scatter_mapbox(houses,
                            lat = 'lat',
                            lon = 'long',
                            hover_name = 'id',
                            size = 'price',
                            color_continuous_scale = px.colors.cyclical.IceFire,
                            size_max = 15,
                            zoom = 10)

    fig.update_layout( mapbox_style ='open-street-map')
    fig.update_layout(height=600, margin={'r':0 ,'t':0 ,'l':0 ,'b':0})
    fig.show()
```

```python
# Create map filters
wd.interactive(update_map, df=fixed(data), price_limit=price_limit, 
                                           living_limit=living_limit,
                                           baths_limit=baths_limit,
                                           base_limit = base_limit,
                                           condition_limit = condition_limit,
                                           const_limit = const_limit)
```

## 5 Adicione os seguintes filtros ao Dashboard:
   - Filtro por data disponível para compra;
   - Filtro por ano de renovovação;
   - Filtro se possui vista para a água ou não.

```python
# Create new columns and defining date format
data['year'] = pd.to_datetime(data['date']).dt.strftime('%Y')
data['date'] = pd.to_datetime(data['date']).dt.strftime('%Y-%m-%d')
data['week'] = pd.to_datetime(data['date']).dt.strftime('%Y-%U')

data.head()
```

```python
# Crate params of filters
date_limit = widgets.SelectionSlider(
    options = data['date'].sort_values().unique().tolist(),
    value='2014-12-01',
    description='Max available date',
    disable=False,
    continuous_update=False,
    style={'description_width':'initial'},
    redout=True)

year_limit = widgets.SelectionSlider(
    options = data['yr_renovated'].sort_values().unique().tolist(),
    value=2000,
    description='Max year',
    disable=False,
    continuous_update=False,
    style={'description_width':'initial'},
    redout=True)

waterfront_limit = widgets.Checkbox(
    options = data['yr_renovated'].sort_values().unique().tolist(),
    value=False,
    description='Is waterfront?',
    disable=False,
    indent=False,
    continuous_update=False,
    style={'description_width':'initial'},
    redout=True)
```

```python
# Define DashBoard function
def update_map(data,date_limit, year_limit, waterfront_limit):
    #filter-data
    df = data[(data['date'] < date_limit) &
              (data['yr_renovated'] >= year_limit) &
              (data['waterfront'] == waterfront_limit)].copy()
    
    fig = plt.figure(figsize=(24,12))
    specs = gs.GridSpec(ncols = 2, nrows = 2, figure=fig)
    
    ax1 = fig.add_subplot(specs[0, :])
    ax2 = fig.add_subplot(specs[1, 0])
    ax3 = fig.add_subplot(specs[1, 1])
    
    #first graph
    by_year = df[['price','year']].groupby('year').sum().reset_index()
    sns.barplot(x='year',y='price', data=by_year, ax=ax1)
    
    
    #second graph
    price_built = df[['price','yr_built']].groupby('yr_built').mean().reset_index()
    sns.lineplot(x='yr_built',y='price', data = price_built, ax = ax2);
    plt.xticks(rotation=90);
    
    #third graph
    by_week = df[['price','week']].groupby('week').mean().reset_index()
    sns.barplot(x='week',y='price', data=by_week, ax=ax3)
    plt.xticks(rotation=90);
```

```python
# Create DashBoard filters
widgets.interactive( update_map,
                     data = fixed(data),
                     date_limit = date_limit,
                     year_limit = year_limit,
                     waterfront_limit = waterfront_limit)
```
